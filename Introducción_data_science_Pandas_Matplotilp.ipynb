# Create visualization

```python
import matplotlib.pyplot as plt
plt.plot('years','califications',data=cal_dict, color='g')

# Configurar títulos
plt.title('Records', color='red')
plt.xlabel('Years')
plt.ylabel('Califications')

# Definir rango en el eje X (de 2020 a 2026)
plt.xlim(2020, 2026)
plt.ylim(0,20)

# Personalizar marcas en el eje X (solo mostrar 2020-2026)
plt.yticks(range(0,21,2))
plt.xticks(range(2020, 2026, 1))  # Muestra cada 1 año
plt.ylim(0, max(cal_dict['califications']) + 2)  # Asegurar que el eje Y comience en 0
plt.show()
```

## Excluded the any columns

```python
df_movies_comuns=df_movies[['title','country','genre','release_year','duration']]
df_movies_comuns.head()
```

## Create un graph of points

```python
fig=plt.figure(figsize=(12,8))
plt.scatter('release_year','duration',data=df_movies_comuns,color='purple')
plt.title('Points Dispertion')
plt.xlabel('years')
plt.ylabel('Durations')

plt.show()
```

## Watch the smallest and largest movies

```python
df_movies_comuns.nlargest(n=1,columns='duration')
```

```python
df_movies_comuns['genre'].value_counts()
```

fit multiple

```python
netflix_df_movies_filtered=df_movies_comuns[(df_movies_comuns['genre']=='Dramas')|(df_movies_comuns['genre']=='Comedies')|(df_movies_comuns['genre']=='Documentaries')]
netflix_df_movies_filtered['genre'].unique()
```

```python
colors=[]
for row,data in netflix_df_movies_filtered.iterrows():
 if data['genre']=='Dramas':
    colors.append('red')
 elif data['genre']=='Comedies':
    colors.append('green')
 else:
    colors.append('blue')
colors

fig=plt.figure(figsize=(12,8))
plt.style.use('fivethirtyeight')
plt.scatter('release_year','duration',data=netflix_df_movies_filtered,color=colors)
plt.title('Points Dispertion')
plt.xlabel('years')
plt.ylabel('Durations')

plt.show()
```

## Bar chart

```python
import seaborn as sns

# Configurar estilo
plt.style.use('fivethirtyeight')

# Contar películas por género
count_movies = df_movies_comuns['genre'].value_counts()

# Generar colores distintos usando Seaborn
colors = sns.color_palette('muted', len(count_movies))

# Crear figura
fig= plt.subplots(figsize=(12, 8))

# Graficar barras con colores distintos
plt.bar(count_movies.index, count_movies.values, color=colors)

# Etiquetas y título
plt.xlabel('Genre films')
plt.ylabel('Amount')
plt.title('Films by Genre', color='b')
plt.xticks(rotation=90)

# Mostrar gráfico
plt.show()
```

## Show the mean duration of movies

```python
df_mean_dutation=df_movies_comuns.groupby('release_year')['duration'].mean().reset_index()
df_mean_dutation
```

```python
# Configurar estilo
plt.style.use('fivethirtyeight')
fig= plt.subplots(figsize=(12, 8))
plt.plot(	'release_year',	'duration',data=df_mean_dutation,color='brown')
plt.show()
```

## Show the most viewed movies in each country.

```python

# Configurar estilo
plt.style.use('fivethirtyeight')
# Crear figura
fig= plt.subplots(figsize=(12, 8))
df_countries=df_movies_comuns['country'].value_counts()
ten_max_countries=df_countries.head(10)
# Generar colores distintos usando Seaborn
colors = sns.color_palette('magma', len(ten_max_countries))
plt.barh(ten_max_countries.index,ten_max_countries.values,color=colors)
plt.xlabel('Amount')
plt.ylabel('Countries')
plt.title('Number of movies for Countries',color='b')
plt.show()
```
